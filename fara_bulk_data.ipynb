# Load and unzip data

```python
import requests
import pandas as pd
import zipfile
import io

folder_path = "fara_files"

# Get data from URL
urls = [
    "https://efile.fara.gov/bulk/zip/FARA_All_Registrants.csv.zip",
    "https://efile.fara.gov/bulk/zip/FARA_All_RegistrantDocs.csv.zip",
    "https://efile.fara.gov/bulk/zip/FARA_All_ShortForms.csv.zip",
    "https://efile.fara.gov/bulk/zip/FARA_All_ForeignPrincipals.csv.zip"
]

for url in urls: 
    response = requests.get(url, verify=False) # disable SSL verification

    # unzip the file and stick it into the fara_files folder
    with zipfile.ZipFile(io.BytesIO(response.content)) as z:
        z.extractall(folder_path)
        print(f"Extracted files to: {folder_path}")
```

# API Call

```python
BASE_URL = "https://efile.fara.gov"
registration_number = "6960"  # Example registration number
endpoint = f"{BASE_URL}/api/v1/ForeignPrincipals/csv/Active/{registration_number}"


response = requests.get(endpoint, verify=False)
# wrap this into a file object so i can pass it to pd.read_csv
file_obj = io.StringIO(response.text)
pd.read_csv(file_obj)
```

```python
endpoint = f"{BASE_URL}/api/v1/Registrants/csv/Active"

response = requests.get(endpoint, verify=False)
# wrap this into a file object so i can pass it to pd.read_csv
file_obj = io.StringIO(response.text)
active = pd.read_csv(file_obj)
```

```python
endpoint = f"{BASE_URL}/api/v1/Registrants/csv/Terminated"

response = requests.get(endpoint, verify=False)
# wrap this into a file object so i can pass it to pd.read_csv
file_obj = io.StringIO(response.text)
terminated = pd.read_csv(file_obj)
```

```python
all_registrants = pd.concat([active, terminated])
all_registrants
```
